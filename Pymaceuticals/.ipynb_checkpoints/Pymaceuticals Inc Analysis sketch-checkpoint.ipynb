```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
clinic = pd.read_csv('raw_data/clinicaltrial_data.csv')
mousedrug = pd.read_csv('raw_data/mouse_drug_data.csv')

len(mousedrug['Mouse ID'].unique())
mousedrug2 = mousedrug.sort_values('Mouse ID').set_index('Mouse ID')
merged_data = clinic.merge(mousedrug, how= 'left', on= 'Mouse ID' )
merged_data.head()
```

## Tumor Response To Treatments

```python
group_tumor = merged_data.groupby(['Drug', 'Timepoint'])
group_drug_time_tumor = group_tumor.agg({'Tumor Volume (mm3)':'mean'})
group_drug_time_tumor
```

```python
Capomulin_data= merged_data[merged_data['Drug'] == 'Capomulin']
Capomulin_data.groupby(['Drug', 'Timepoint']).agg({'Tumor Volume (mm3)':'mean'})
```

```python
def tumor_overtime_treament():
    y = merged_data[['Timepoint', 'Tumor Volume (mm3)','Drug']]
    time_drug_volums = y.groupby(['Timepoint', 'Drug']).agg({'Tumor Volume (mm3)':'sem'})
    time_drug_volums= time_drug_volums.unstack()
    return time_drug_volums
tumor_overtime_treament().head()
```

```python
data3= tumor_overtime_treament().ix[(0,5,10,15,10,15,20,25,30,35,40,45), 'Tumor Volume (mm3)']
data3
listtt = data3.columns.tolist()
listt3 = []
for i in listtt :
    change_deci =(data3.loc[45,i] - data3.loc[5,i])/data3.loc[5,i] 
    listt3.append(change_deci)
change_overtime = pd.DataFrame(listt3, listtt).ix[['Capomulin','Infubinol', 'Ketapril', 'Placebo' ]]
change_overtime[0].tolist()
```

```python


plot_treament= tumor_overtime_treament().ix[(0,5,10,15,10,15,20,25,30,35,40,45), 'Tumor Volume (mm3)']
x = (0,5,10,15,10,15,20,25,30,35,40,45)
fig1 = plt.figure(figsize = (10,8))
plt.errorbar(x,plot_treament['Capomulin'],yerr= -0.19475302667894173 ,  fmt= 'o--', c = 'b', label ='Capomulin', elinewidth = 1, capthick = 1, capsize = 5)
plt.errorbar(x,plot_treament['Infubinol'], yerr= 0.46123471727851845,fmt= 'o--', c= 'r', label = 'Infubinol', elinewidth = 1, capthick = 1, capsize = 5 )
plt.errorbar(x,plot_treament['Ketapril'], yerr= 0.5702879468660608, fmt= 'o--', c= 'g', label ='Ketapril', elinewidth = 1, capthick = 1, capsize = 5 )
plt.errorbar(x,plot_treament['Placebo'],yerr= 0.5129796048315153, fmt= 'o--', c ='k', label = 'Placebo', elinewidth = 1, capthick = 1, capsize = 5)
plt.ylim(20,80)
plt.xlim(0,50)
plt.legend(loc = 'upper left')
plt.grid(alpha = 0.3)
plt.xlabel('Time(Days)')
plt.ylabel('Tumor Volume(mm3)')
plt.title('Tumor Response to Treatment', fontsize = 18)

plt.show()
```

```python

y = pd.DataFrame(pd.Series(listt3, listtt))
plt.style.use('seaborn-notebook')
fig = plt.figure(figsize=(9,8))
ax = y.ix[['Capomulin', 'Infubinol', 'Ketapril','Placebo' ], 0].plot(kind = 'bar', colors = ['g', 'r','r','r'])
plt.grid(alpha = 0.4, linestyle = '--')
plt.xlabel('Treatment')
plt.ylabel('% Change Volumns Overtimes')
plt.title("Tumer % Change over 45 Day Treatment")
plt.ylim(-0.2, 0.8)

x_axis = np.arange(len(['Capomulin', 'Infubinol', 'Ketapril','Placebo' ]))
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations,['Capomulin', 'Infubinol', 'Ketapril','Placebo' ])

rects = ax.patches

# For each bar: Place a label
for rect in rects:
    # Get X and Y placement of label from rect
    y_value = rect.get_height()
    x_value = rect.get_x() + rect.get_width() / 2

    # Number of points between bar and label. Change to your liking.
    space = 5
    # Vertical alignment for positive values
    va = 'bottom'

    # If value of bar is negative: Place label below bar
    if y_value < 0:
        # Invert space to place label below
        space *= -1
        # Vertically align label at top
        va = 'top'

    # Use Y value as label and format number with one decimal place
    label = "{:.2%}".format(y_value)

    # Create annotation
    plt.annotate(
        label,                      # Use `label` as label
        (x_value, y_value),         # Place label at end of the bar
        xytext=(0, space),          # Vertically shift label by `space`
        textcoords="offset points", # Interpret `xytext` as offset in points
        ha='center',                # Horizontally center label
        va=va) 
plt.axhline(y=0, color='black', linestyle='-')
plt.show()
```
